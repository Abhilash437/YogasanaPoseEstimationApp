```python
def calculate_angle(a,b,c):
    a = np.array(a) # First
    b = np.array(b) # Mid
    c = np.array(c) # End
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle >180.0:
        angle = 360-angle
        
    return angle 
```

```python
def tad_eval(user_pose):
    #compare with original
    diff={}
    for x in user_pose:
        diff[x]=abs(user_pose[x]-t[x])
    # print(diff)
    return (diff)
    #make a funcion for each step
```

```python
def vrik_eval(user_pose):
    #compare with original
    diff={}
    for x in user_pose:
        diff[x]=abs(user_pose[x]-v[x])
    # print(diff)
    return (diff)
```

```python
def bhuj_eval(user_pose):
    #compare with original
    diff={}
    for x in user_pose:
        diff[x]=abs(user_pose[x]-b[x])
    # print(diff)
    return (diff)
```

```python
def t_1(diff):
    all=True
    angles=['r1','r2','l1','l2']
    for angle in angles:
        if(diff[angle]>5):
            all=False
    if(all):    
        return [feedbacks['tadasana'][1]['Good'],True]
    else:
        return [feedbacks['tadasana'][1]['Bad'],False]
```

```python
def t_2(coordinates):
    cond=((coordinates['left_shoulder']>=coordinates['left_elbow'])and (coordinates['right_shoulder']>=coordinates['right_elbow']))
    if(cond):
        return [feedbacks['tadasana'][2]['Good'], True]
    else:
        return [feedbacks['tadasana'][2]['Bad'],False]
```

```python
def t_3(diff):
    all=True
    angles=['r3','l3','r4','l4']
    for angle in angles:
        if(diff[angle]>10):
            all=False
    if(all):
        return [feedbacks['tadasana'][3]['Good'], True]
    else:
        return [feedbacks['tadasana'][3]['Bad'], False]
```

```python
import cv2
import mediapipe as mp
import time
#import pygame.mixer

count = 1
asanas = ["Bhujangaasana", "Vrikshasana", "Tadasana"]

# Initialize MediaPipe hands module
mp_hands = mp.solutions.hands
hands = mp_hands.Hands()

# Initialize MediaPipe drawing module
mp_drawing = mp.solutions.drawing_utils

# Capture video from the camera
cap = cv2.VideoCapture(0)

# Initialize variables for tracking finger raise time
i_finger_raised = False
m_finger_raised = False
r_finger_raised = False
l_finger_raised = False
t_finger_raised = False

i_start_time = None
t_start_time = None
m_start_time = None
r_start_time = None
l_start_time = None


while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        continue

    # Convert the BGR image to RGB
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Process the image with MediaPipe hands model
    results = hands.process(frame_rgb)

    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            # Determine if the hand is visible (within the frame)
            is_visible = all(0.0 <= landmark.x <= 1.0 and 0.0 <= landmark.y <= 1.0 for landmark in hand_landmarks.landmark)

            if is_visible:
                # Calculate the distance between the index finger tip and the base of the index finger
                index_finger_tip = [hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_DIP].x,hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_DIP].y]
                index_finger_base = [hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_MCP].x,hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_MCP].y]
                index_finger_mid= [hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_PIP].x,hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_PIP].y]

                middle_finger_tip = [hand_landmarks.landmark[11].x,hand_landmarks.landmark[11].y]
                middle_finger_base = [hand_landmarks.landmark[9].x,hand_landmarks.landmark[9].y]
                middle_finger_mid = [hand_landmarks.landmark[10].x,hand_landmarks.landmark[10].y]

                ring_finger_tip = [hand_landmarks.landmark[15].x, hand_landmarks.landmark[15].y]
                ring_finger_base =[hand_landmarks.landmark[13].x, hand_landmarks.landmark[13].y]
                ring_finger_mid = [hand_landmarks.landmark[14].x, hand_landmarks.landmark[14].y]

                
                pinky_finger_tip = [hand_landmarks.landmark[19].x, hand_landmarks.landmark[19].y]
                pinky_finger_base = [hand_landmarks.landmark[17].x, hand_landmarks.landmark[17].y]
                pinky_finger_mid = [hand_landmarks.landmark[18].x, hand_landmarks.landmark[18].y]

                
                thumb_finger_tip = [hand_landmarks.landmark[4].x, hand_landmarks.landmark[4].y]
                thumb_finger_mid = [hand_landmarks.landmark[3].x, hand_landmarks.landmark[3].y]
                thumb_finger_base = [hand_landmarks.landmark[2].x,hand_landmarks.landmark[2].y]


                if calculate_angle(index_finger_tip,index_finger_mid,index_finger_base)>165:  # You may need to adjust this threshold
                    if not i_finger_raised:
                        i_start_time = time.time()
                    i_finger_raised = True
                else:
                    i_finger_raised = False

                if calculate_angle(middle_finger_tip,middle_finger_mid,middle_finger_base)>165:  # You may need to adjust this threshold
                    if not m_finger_raised:
                        m_start_time = time.time()
                    m_finger_raised = True
                else:
                    m_finger_raised = False
                
                if calculate_angle(ring_finger_tip,ring_finger_mid,ring_finger_base)>165:  # You may need to adjust this threshold
                    if not r_finger_raised:
                        r_start_time = time.time()
                    r_finger_raised = True
                else:
                    r_finger_raised = False

                if calculate_angle(thumb_finger_tip,thumb_finger_mid,thumb_finger_base)>165:  # You may need to adjust this threshold
                    if not t_finger_raised:
                        t_start_time = time.time()
                    t_finger_raised = True
                else:
                    t_finger_raised = False
                
                if calculate_angle(pinky_finger_tip,pinky_finger_mid,pinky_finger_base)>165:  # You may need to adjust this threshold
                    if not l_finger_raised:
                        l_start_time = time.time()
                    l_finger_raised = True
                else:
                    l_finger_raised = False

                # Draw hand landmarks
                mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)

    # Check if the index finger has been raised continuously for more than one second
    
    if i_finger_raised and m_finger_raised and r_finger_raised and l_finger_raised and t_finger_raised \
    and time.time() - i_start_time >= 1 and time.time()-m_start_time>=1 and time.time() - r_start_time >= 1 \
    and time.time()-l_start_time>=1 \
    and time.time() - t_start_time >= 1:
        asana=asanas[count][0].lower()
        cv2.putText(frame, asana, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

        # i_finger_raised=False
        # m_finger_raised=False
        # r_finger_raised=False
        # l_finger_raised=False
        # t_finger_raised=False
        break
        
        
    elif i_finger_raised and m_finger_raised and time.time() - i_start_time >= 2 and time.time()-m_start_time>=2:
        count = max(0,(count - 1)%3)
        cv2.putText(frame, asanas[count], (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
        # tts=gTTS(asanas[count])
        # tts.save("output.mp3")
        # os.system("afplay output.mp3")
        i_finger_raised=False
        m_finger_raised=False
        tts=gTTS(asanas[count])
        tts.save("output.mp3")
        os.system("afplay output.mp3")
        #playsound('output.mp3')
#         pygame.mixer.init()

# # Load the audio file (change 'your_audio_file.mp3' to your actual file)
#         pygame.mixer.music.load('output.mp3')

#         # Play the audio
#         pygame.mixer.music.play()

#         # Keep the program running while the audio plays
#         # while pygame.mixer.music.get_busy():
#         #     pass
    
    elif i_finger_raised and time.time() - i_start_time >= 2:
        # cv2.putText(frame, "Index Finger Raised for 1s", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
        count = (count + 1)%3
        cv2.putText(frame, asanas[count], (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
        i_finger_raised=False
        tts=gTTS(asanas[count])
        tts.save("output.mp3")
        os.system("afplay output.mp3")
    # else:
    #     cv2.putText(frame,"Perform gesture", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
    

    # Display the result
    cv2.imshow('Hand Detection', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object and close OpenCV windows
cap.release()
cv2.destroyAllWindows()

# Release the MediaPipe hands model
hands.close()
```
